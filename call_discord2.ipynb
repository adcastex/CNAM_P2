```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_excel("data/Extraction_IJ/Extraction_IJ_hebdo_UNIV_LYON_Compte U.xlsx", sheet_name='Extraction hebdomadaire 1')
df2 = pd.read_excel("data/Extraction_IJ/Extraction_IJ_hebdo_UNIV_LYON_Compte U.xlsx", sheet_name='Extraction hebdomadaire 2')
```

```python
result_df = pd.concat([df, df2], axis=0)
```

```python
result_df
```

```python
resultat_groupby = result_df.groupby(['Code semaine'])['Délai après réception de la déclaration de salaire'].mean().reset_index()
```

```python
resultat_groupby
```

```python
resultat_groupby = resultat_groupby.dropna()
```

```python
resultat_groupby
```

```python
df_tx = pd.read_excel("data/Projet 2_IJ et front_indicateurs V3_2.xlsx", sheet_name='Tx itération multicanal',skiprows=range(1, 21),header=1)
```

```python
df_tx
```

```python
resultat_groupby2 = df_tx.groupby(['Période'])['Mesure 3'].mean().reset_index()
```

```python
resultat_groupby2
```

```python
result_join = pd.merge(resultat_groupby, resultat_groupby2, left_on='Code semaine', right_on='Période', how='left')
# Drop the duplicate 'Num organisme' column if needed
# result_join = result_join.drop(columns='Num organisme')
result_join = result_join.rename(columns={'Mesure 3': 'Solicitation front'})
```
