```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la colonne 'Date prescription' en format datetime
df['Date prescription'] = pd.to_datetime(df['Date prescription'], format='%d/%m/%Y')

# Extraire le mois à partir de la colonne 'Date prescription' et convertir en str
df['Mois'] = df['Date prescription'].dt.to_period('M').astype(str)

# Créer un graphique d'évolution mensuelle du nombre de PRN par organisme
plt.figure(figsize=(10, 6))
for organisme, data in df.groupby('Organisme'):
    plt.plot(data['Mois'], data['Nombre PRN'], label=f'Organisme {organisme}', marker='o')

# Ajouter des labels et une légende
plt.xlabel('Mois')
plt.ylabel('Nombre PRN')
plt.title('Évolution mensuelle du nombre PRN par organisme')
plt.legend(title='Organisme', bbox_to_anchor=(1, 1), loc='upper left')

# Afficher le graphique
plt.show()
```
